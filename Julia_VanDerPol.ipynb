```python
using PyPlot
```

```python
function DY(t,X,ɛ=5)
  
  x=X[1]
  y=X[2]
  dx=y;
  dy=-x+ (1-x^2)y*ɛ
  return [dx;dy]
end
```

```python
function RK4(f,t, Y,h)
  k1=h*f(t,Y)
  k2=h*f(t+0.5h, Y +0.5k1)
  k3=h*f(t+0.5h, Y +0.5k2)
  k4=h*f(t+h,Y+k3)
  ynext=Y + (1/6.0 * k1)+ (1/3.0 * k2) + (1/3 * k3) + 1/6*k4
  return ynext
end
```

```python
Tspan=[0,20]
X0=[-2;-2]
h2=0.1
iterations=int(floor(Tspan[2]/h2))
Y=copy(X0)
t=0
Out=zeros(2,iterations)
@time for i=1:iterations
  Y=RK4(DY,t,Y,h)
  t=t+h
  Out[:,i]=Y
end
```

```python
h=0.05
iterations=int(floor(Tspan[2]/h))
t=0
Y=copy(X0)
Out2=zeros(2,iterations)
@time for i=1:iterations
  Y=RK4(DY,t,Y,h2)
  t=t+h2
  Out2[:,i]=Y
end
```

```python
scatter(Out[1,:],Out[2,:],marker="x",label="RK4,h=$h")
scatter(Out2[1,:],Out2[2,:],marker="o",label="RK4,h=$h2")
plot(X0[1],X0[2],marker="o", markerfacecolor="red", markersize=10, label="Starting Point")
legend(loc="upper center", bbox_to_anchor=(0.5, 1.15),
          ncol=3, fancybox=true, shadow=true)
show()
```
